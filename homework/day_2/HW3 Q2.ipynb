```python
import numpy as np
```

```python
# Define a function
# Define the number of particles and the size of the box
def random_particle(number, box_length):
    sample_coords = [] # Create a coordinate
    r = box_length/2   # Periodic boundary restriction
    for i in range(number):
        particle = []# Create a list of xyz coordinates of each particle
        x = np.random.uniform(-r, r)
        y = np.random.uniform(-r, r)
        z = np.random.uniform(-r, r)
        particle = [x, y, z]# output as [x,y,z]
        sample_coords.append(particle) # Add the particle into the aboved 2-Dimensional list
    
    return (number, sample_coords)
```

```python
number, sample_coords = random_particle(500, 10)
```
